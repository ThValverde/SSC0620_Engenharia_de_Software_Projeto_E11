```python
from dataclasses import dataclass

@dataclass
class Usuario:
    id_usuario: int
    nome_completo: str
    email: str
    senha_criptografada: str  
    data_cadastro: str
    ativo: bool

@dataclass
class Administrador(Usuario):
    eh_superusuario: bool = True

@dataclass
class FuncionarioSecretaria(Usuario):
    matricula: str
    setor: str

@dataclass
class ResponsavelEstabelecimento(Usuario):
    cpf: str
    telefone_comercial: str
    estabelecimentos_gerenciados: list[estabelecimento]    
```
